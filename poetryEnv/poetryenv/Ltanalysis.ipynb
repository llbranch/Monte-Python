```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import find_peaks, peak_prominences, peak_widths
```

```python
"""Pull the entire data set"""
data = pd.read_csv('monte_carlo_outputTest.txt', sep="\t")
data.columns =['Time', 'V(compout)', 'V(input)', 'V(out)', 'V(ref)', 'I(comp)']
# data.to_numpy() 
# data.describe()
```

```python
"""See a Response"""

# Convert to Data Frame 
df = pd.DataFrame(data)
print("Data frame Types:")
print(df.dtypes)

# truncate data to better see the repsonse 
newdf = df.truncate(after=605)

# Define arrays

y1 = newdf["V(input)"]
y2 = newdf["V(compout)"]
y3 = newdf["V(ref)"]
x = newdf["Time"]

print(type(x))

"""Plot"""

fig, ax = plt.subplots()

ax.plot(x,y1, 'g-', label='V(Input)')
ax.plot(x,y2, 'b-', label='V(compout)')
ax.plot(x,y3, 'r-', label='V(ref)')
plt.title('Comparator Response')
plt.ylabel("Voltage")
plt.xlabel("Time (ns)")
plt.legend()
newdf.plot(y="V(compout)", x ="Time", title="Response")
```

```python
"""Peaks"""
peaks, _ = find_peaks(y2.to_numpy(), distance=5)
VComp = y2.to_numpy()
Vpeak = VComp[peaks]
print(Vpeak)
```
