```python
""" Python program to show time by perf_counter() """ 
from time import perf_counter, perf_counter_ns

# printing the time
start = perf_counter()
print(f'{perf_counter()} s')
end = perf_counter() - start
print(f'Time to call perf counter and print (s): ', end)

start = perf_counter()
print(f'{perf_counter_ns()} ns')
end = perf_counter() - start
print(f'Time to call perf counter ns and print (ns): ', end)
```

```python
""" Plot the hist """
from time import perf_counter, perf_counter_ns
import numpy as np
import matplotlib.pyplot as plt

# Create an array of samples 
counter = []
for _ in range(500_000): 
    start = perf_counter()
    call = perf_counter()
    end = perf_counter() - start
    end = end*10e9
    counter.append(end)


# Create an array of samples 
counter_ns = []
for _ in range(500_000): 
    start = perf_counter_ns()
    call = perf_counter_ns()
    end = perf_counter_ns() - start
    counter_ns.append(end)

# print(counter[0:4])
# print(counter_ns[0:4])
counter_av = np.average(counter)
counter_std = np.std(counter)

counter_ns_av = np.average(counter_ns)
counter_ns_std = np.std(counter_ns)

print(f'Counter\t\tAverage: {counter_av}\tSTD: {counter_std}')
print(f'Counter_ns\tAverage: {counter_ns_av}\t\tSTD: {counter_ns_std}')

plt.hist(counter, 60, range=[0,1e4], label='counter', color = 'C1')
plt.hist(counter_ns, 60, range=[0,1e4], label='perf_counter_ns', color = 'C2')
plt.xlabel('Time (ns)')
plt.ylabel('Counts')
plt.title('perf_counter vs perf_counter_ns: Execution times 500k samples')
plt.grid()
plt.legend(loc='best', fontsize=14)
plt.show()
```

```python
from time import perf_counter, perf_counter_ns
import numpy as np
import matplotlib.pyplot as plt

DATA = []

for _ in range( 100_000 ) :

    start = perf_counter()
    # print(f'{perf_counter()} s')
    call = perf_counter()
    end = perf_counter() - start
    DATA.append( end )

DATA2 = []

for _ in range( 100_000 ) :

    start = perf_counter()
    # print(f'{perf_counter_ns()} ns')
    call = perf_counter_ns()
    end = perf_counter() - start
    DATA2.append( end )

plt.hist( DATA , 100 , range = [0,1e-6] )
plt.hist( DATA2 , 100 , range = [0,1e-6] )

plt.show()
```
