```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Analyzing the timing behavior of my simulation Using 3 of my 4 Core CPU

Need to take 5 samples per run to get a good fit --> Todo: Take more samples

```python
df = pd.DataFrame({'particle_count': [1,5,10,15,20], 'time': [5.069007,22.842866,43.660468,69.416799,95.166945]})
df2 = pd.DataFrame({'particle_count': [1,5,5,10,15,20], 'time': [5.589515,24.383691,20.166244,57.119575,69.284174,84.507807]})
df = pd.concat([df,df2])
df.describe()
```

```python
fig0, ax0 = plt.subplots()
ax0.plot(df['particle_count'],df['time'], marker='.', ls='none', label='Simulation Times')
m,b = np.polyfit(df['particle_count'],df['time'], 1)
ax0.plot(df['particle_count'],m*df['particle_count']+b, marker='.',ls='-.', alpha=0.6, label=f'Fitted Line\n y={m:.1f}x+{b:.1f}')
ax0.set_xlabel('# of Particles')
ax0.set_ylabel('time [s]')
ax0.set_title('time vs. # of particles simulated')
ax0.grid()
ax0.legend()
plt.show()
```
